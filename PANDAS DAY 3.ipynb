```python
# merge operations
# with same key column name as empid
data4 = {
    'empid':[101,102,103,104],
    'salary':[204,456,678,345],
    'pf':[234,567,211,567],
    
}

data5 = {
    'empid':[101,102,104,105],
    'mobno':[2044,4564,6784,3454],
    'houseno':[65,64,34,56] 
}
```

```python
df4 = pd.DataFrame(data4)
df4
```

```python
df5 = pd.DataFrame(data5)
df5
```

```python
# default merge type is inner
pd.merge(df4,df5)
```

```python
# take everything from left DF and common one from right table
pd.merge(df4,df5,how='left')
```

```python
# take everything from right DF and common one from left table
pd.merge(df4,df5,how='right')
```

```python
# union operation
pd.merge(df4,df5,how='outer')
```

```python
pd.merge(df4,df5,how='right',on='empid')
```

```python
# with diff key column name as empid1 and empid2
data6 = {
    'empid1':[101,102,103,104],
    'salary':[204,456,678,345],
    'pf':[234,567,211,567],
    
}

data7 = {
    'empid2':[101,102,104,105],
    'mobno':[2044,4564,6784,3454],
    'houseno':[65,64,34,56] 
}
```

```python
df6=pd.DataFrame(data6)
df6
```

```python
df7=pd.DataFrame(data7)
df7
```

```python
# with diff key column name as empid1 and empid2
pd.merge(df6,df7,left_on='empid1',right_on='empid2')
```

```python
df_sales = pd.read_csv(r"C:\Users\udaya\Desktop\Python\DATA_SCIENCE\PANDAS\DATASET\sales_data_final.csv")
df_sales.head()
```

```python
def profit_flag(a):
    if a>0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(2)
```

```python
# apply function - used to apply specific function in column level
df_sales['profit_flag'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales['profit_flag'].value_counts()
```

```python
# passing inbuild function in apply
# used only one column in apply
df_sales['len_cust_name']=df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['quantity_flag'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.tail()
```

```python
# using lambda function in apply

df_sales['quantity_square'] = df_sales['quantity'].apply(lambda a: a**2)
```
