```python
import cufflinks as cf
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# cufflinks go offline
cf.go_offline()
```

```python
data = np.random.randn(100,3)
```

```python
df = pd.DataFrame(data,columns=['A','B','C'])
df.head()
```

```python
df.iplot()
# x axis in below graph will be the row index of DF
```

```python
df1 = pd.DataFrame(np.random.randint(3,100,(100,1)))
```

```python
df1.head()
```

```python
df1.iplot()
```

```python
# by default iplot() in DF will consider row index as x axis and value in y-axis
# if we need to make x-axis and y-axis as a column values
df.iplot(x='A',y='B',mode='markers')
```
